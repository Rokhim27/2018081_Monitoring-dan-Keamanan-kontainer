```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score
```

```python
np.random.seed(42)

data = []

for i in range(100):
    R = np.random.randint(50, 255)
    G = np.random.randint(50, 255)
    B = np.random.randint(50, 255)

    # aturan sederhana label
    if G > R and G > B:
        label = "Sehat"
    elif R > G:
        label = "Coklat"
    else:
        label = "Kuning"

    data.append([R, G, B, label])

df = pd.DataFrame(data, columns=['R_mean','G_mean','B_mean','Label'])

# simpan dataset
df.to_csv("dataset_daun.csv", index=False)

print(df.head())
```

```python
X = df[['R_mean','G_mean','B_mean']]
y = df['Label']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
knn = KNeighborsClassifier(n_neighbors=3)
knn.fit(X_train, y_train)
```

```python
y_pred = knn.predict(X_test)
```

```python
acc = accuracy_score(y_test, y_pred)
print("Akurasi:", acc)
```

```python
contoh = [[120, 180, 90]]
prediksi = knn.predict(contoh)

print("Prediksi daun:", prediksi[0])
```
